                    # Model Development Evaluation and Testing

```python
# Separating the features (independent variables) and the target variable (dependent variable)

# Creating the feature set by dropping the 'FWI' column from the DataFrame
X = df.drop(['FWI'], axis=1)  # X contains all columns except 'FWI', which will be our target variable

# Assigning the target variable to Y
Y = df['FWI']  # Y contains only the 'FWI' column, representing the dependent variable we want to predict
```

```python
from sklearn.model_selection import train_test_split  # Importing the train_test_split function

# Splitting the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(
    X,        # The features set (independent variables)
    Y,        # The target variable (dependent variable)
    test_size=0.2,  # Proportion of the dataset to include in the test split (20%)
    random_state=42  # Seed for random number generator to ensure reproducibility
)
```

```python
# Initializing the scaler
scaler = StandardScaler()  # Create an instance of StandardScaler to standardize the feature values

# Scaling the features
x_train_scaled = scaler.fit_transform(x_train)  # Fit the scaler on the training data and transform it
x_test_scaled = scaler.transform(x_test)  # Transform the testing data using the already fitted scaler
```

            # Multiple Linear Regression with cross validation

```python
# Initializing the linear regression model
linear_model = LinearRegression()  # Create an instance of the LinearRegression model

# Fitting the model to the scaled training data
linear_model.fit(x_train_scaled, y_train)  # Train the model using the scaled training features and target values

# Predictions
y_train_pred = linear_model.predict(x_train_scaled)  # Predict the target values for the training set
y_test_pred = linear_model.predict(x_test_scaled)  # Predict the target values for the testing set

# Cross-Validation
cv_scores = cross_val_score(linear_model, x_train_scaled, y_train, cv=5)  # Perform 5-fold cross-validation
print("Linear Regression - CV Mean Score:", np.mean(cv_scores))  # Print the mean cross-validation score

# Print evaluation metrics for the training set
print("Linear Regression - Train MSE:", mean_squared_error(y_train, y_train_pred),  # Calculate Mean Squared Error (MSE) for training set
      "Train RÂ²:", r2_score(y_train, y_train_pred),  # Calculate R-squared for training set
      "Train MAE:", mean_absolute_error(y_train, y_train_pred))  # Calculate Mean Absolute Error (MAE) for training set

# Print evaluation metrics for the testing set
print("Linear Regression - Test MSE:", mean_squared_error(y_test, y_test_pred),  # Calculate MSE for testing set
      "Test RÂ²:", r2_score(y_test, y_test_pred),  # Calculate R-squared for testing set
      "Test MAE:", mean_absolute_error(y_test, y_test_pred))  # Calculate MAE for testing set
```

                       # Polynomial Regression with Cross-Validation

```python
# Creating Polynomial Features
poly_features = PolynomialFeatures(degree=2)  # Instantiate PolynomialFeatures with a degree of 2
x_train_poly = poly_features.fit_transform(x_train_scaled)  # Transform the scaled training features to polynomial features
x_test_poly = poly_features.transform(x_test_scaled)  # Transform the scaled testing features to polynomial features

# Fitting the Polynomial Regression model
poly_model = LinearRegression()  # Create an instance of the LinearRegression model for polynomial regression
poly_model.fit(x_train_poly, y_train)  # Train the polynomial model using the polynomial features and the target values

# Predictions
y_train_poly_pred = poly_model.predict(x_train_poly)  # Predict the target values for the training set using the polynomial model
y_test_poly_pred = poly_model.predict(x_test_poly)  # Predict the target values for the testing set using the polynomial model

# Cross-Validation
cv_poly_scores = cross_val_score(poly_model, x_train_poly, y_train, cv=5)  # Perform 5-fold cross-validation on the polynomial model
print("Polynomial Regression - CV Mean Score:", np.mean(cv_poly_scores))  # Print the mean cross-validation score

# Print evaluation metrics for the training set
print("Polynomial Regression - Train MSE:", mean_squared_error(y_train, y_train_poly_pred),  # Calculate MSE for training set
      "Train RÂ²:", r2_score(y_train, y_train_poly_pred))  # Calculate R-squared for training set

# Print evaluation metrics for the testing set
print("Polynomial Regression - Test MSE:", mean_squared_error(y_test, y_test_poly_pred),  # Calculate MSE for testing set
      "Test RÂ²:", r2_score(y_test, y_test_poly_pred))  # Calculate R-squared for testing set
```

            #  Regularization Models (Lasso and Ridge) with Cross-Validation

              # Fit, Evaluate, and Cross-Validate Lasso Regression

```python
# Initializing the Lasso Regression model
lasso_model = Lasso()  # Create an instance of the Lasso regression model

# Defining the parameter grid for hyperparameter tuning
lasso_params = {'alpha': np.logspace(-4, 4, 10)}  # Set a range of alpha values from 10^-4 to 10^4 on a logarithmic scale

# Setting up GridSearchCV for hyperparameter tuning with cross-validation
lasso_grid = GridSearchCV(lasso_model, lasso_params, cv=5)  # Initialize GridSearchCV with the Lasso model and parameters, using 5-fold cross-validation

# Fitting the GridSearchCV object to the training data
lasso_grid.fit(x_train_scaled, y_train)  # Train the model using the scaled training data

# Retrieving and printing the best alpha value found during the grid search
print("Best Lasso Alpha:", lasso_grid.best_params_['alpha'])  # Display the best alpha parameter from the grid search

# Making predictions on the test set using the best estimator from grid search
y_test_lasso_pred = lasso_grid.predict(x_test_scaled)  # Predict the target values for the scaled test set

# Performing cross-validation on the best Lasso model
lasso_cv_scores = cross_val_score(lasso_grid.best_estimator_, x_train_scaled, y_train, cv=5)  # Compute cross-validation scores

# Printing the mean cross-validation score for the Lasso model
print("Lasso Regression - CV Mean Score:", np.mean(lasso_cv_scores))  # Display the mean cross-validation score

# Printing evaluation metrics for the test set
print("Lasso Regression - Test MSE:", mean_squared_error(y_test, y_test_lasso_pred),  # Calculate and print the MSE for the test set
      "Test RÂ²:", r2_score(y_test, y_test_lasso_pred))  # Calculate and print the R-squared value for the test set
```

                # Fit, Evaluate, and Cross-Validate Ridge Regression

```python
# Initializing the Ridge Regression model
ridge_model = Ridge()  # Create an instance of the Ridge regression model

# Defining the parameter grid for hyperparameter tuning
ridge_params = {'alpha': np.logspace(-4, 4, 10)}  # Set a range of alpha values from 10^-4 to 10^4 on a logarithmic scale

# Setting up GridSearchCV for hyperparameter tuning with cross-validation
ridge_grid = GridSearchCV(ridge_model, ridge_params, cv=5)  # Initialize GridSearchCV with the Ridge model and parameters, using 5-fold cross-validation

# Fitting the GridSearchCV object to the training data
ridge_grid.fit(x_train_scaled, y_train)  # Train the model using the scaled training data

# Retrieving and print the best alpha value found during the grid search
print("Best Ridge Alpha:", ridge_grid.best_params_['alpha'])  # Display the best alpha parameter from the grid search

# Making predictions on the test set using the best estimator from grid search
y_test_ridge_pred = ridge_grid.predict(x_test_scaled)  # Predict the target values for the scaled test set

# Performing cross-validation on the best Ridge model
ridge_cv_scores = cross_val_score(ridge_grid.best_estimator_, x_train_scaled, y_train, cv=5)  # Compute cross-validation scores

# Printing the mean cross-validation score for the Ridge model
print("Ridge Regression - CV Mean Score:", np.mean(ridge_cv_scores))  # Display the mean cross-validation score

# Printing evaluation metrics for the test set
print("Ridge Regression - Test MSE:", mean_squared_error(y_test, y_test_ridge_pred),  # Calculate and print the MSE for the test set
      "Test RÂ²:", r2_score(y_test, y_test_ridge_pred))  # Calculate and print the R-squared value for the test set
```

                # Saving and Testing the Model

```python
import pickle  # Importing the pickle module for serializing objects

# Saving the Standard Scaler to a file
scaler_filename = 'scaler.pkl'  # Define the filename for the scaler
with open(scaler_filename, 'wb') as file:  # Open the file in write-binary mode
    pickle.dump(scaler, file)  # Serialize the scaler object and save it to the file

# Saving the Linear Regression model to a file
linreg_filename = 'linear_model.pkl'  # Define the filename for the linear regression model
with open(linreg_filename, 'wb') as file:  # Open the file in write-binary mode
    pickle.dump(linear_model, file)  # Serialize the linear regression model and save it to the file

# Saving the Lasso Regression model to a file
lasso_filename = 'lasso_model.pkl'  # Define the filename for the Lasso regression model
with open(lasso_filename, 'wb') as file:  # Open the file in write-binary mode
    pickle.dump(lasso_grid, file)  # Serialize the Lasso regression model (including its best estimator) and save it to the file

# Saving the Ridge Regression model to a file
ridge_filename = 'ridge_model.pkl'  # Define the filename for the Ridge regression model
with open(ridge_filename, 'wb') as file:  # Open the file in write-binary mode
    pickle.dump(ridge_grid, file)  # Serialize the Ridge regression model (including its best estimator) and save it to the file
```

```python
# Loading the Standard Scaler from the file
scaler_filename = 'scaler.pkl'  # Define the filename for the scaler
with open(scaler_filename, 'rb') as file:  # Open the file in read-binary mode
    loaded_scaler = pickle.load(file)  # Deserialize the scaler object from the file

# Loading the Linear Regression model from the file
linreg_filename = 'linear_model.pkl'  # Define the filename for the linear regression model
with open(linreg_filename, 'rb') as file:  # Open the file in read-binary mode
    loaded_linear_model = pickle.load(file)  # Deserialize the linear regression model from the file

# Loading the Lasso Regression model from the file
lasso_filename = 'lasso_model.pkl'  # Define the filename for the Lasso regression model
with open(lasso_filename, 'rb') as file:  # Open the file in read-binary mode
    loaded_lasso_model = pickle.load(file)  # Deserialize the Lasso regression model from the file

# Loading the Ridge Regression model from the file
ridge_filename = 'ridge_model.pkl'  # Define the filename for the Ridge regression model
with open(ridge_filename, 'rb') as file:  # Open the file in read-binary mode
    loaded_ridge_model = pickle.load(file)  # Deserialize the Ridge regression model from the file
```
